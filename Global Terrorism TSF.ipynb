```python
import pandas as pd  # import the library

pd.set_option('display.max_columns',None)  # this displays the maximum columns in the dataframe
```

```python
df = pd.read_csv(r"C:\Users\hp\Desktop\TSF-Internship\globalterrorism.csv", encoding="ISO-8859-1")
df.head()                      
```

```python
# Shows the total rows and columns of the data

df.shape
```

```python
# This list all column names of the data

df.columns.to_list()
```

```python
# columns that will be used in the analysis

df = df[["iyear","country_txt","region_txt","provstate",
         "suicide","attacktype1_txt","targtype1_txt",
          "gname","weaptype1_txt","nkill","nwound"]]
df.head()
```

```python
#Renaming the columns

df.rename(columns={"iyear":"Year","country_txt":"Country",
                   "region_txt":"Region","provstate":"State",
                   "attacktype1_txt":"Type of Attack","targtype1_txt":"Target Type",
                   "gname":"Name of Group","weaptype1_txt":"Type of Weapon",
                   "nkill":"Killed","nwound":"Wounded", 'suicide':'Suicide'},inplace=True)
```

```python
# Provides a concise summary of the dataframe

df.info()
```

```python
# Replace null values with 0

df['Killed'] = df['Killed'].fillna(0)
df['Wounded'] = df['Wounded'].fillna(0)
```

```python
# Add a column for casualty

df['Casualty'] = df['Killed'] + df['Wounded']
df.head()
```

```python
# Check for missing values

df.isnull().sum()
```

```python
# Provides a statistical summary of the data

df.describe()
```
